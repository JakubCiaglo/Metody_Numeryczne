# Zadanie 1

```python
def z1(x):
    return (6-2*x)/(6+4*x+x**2)

def z2(x):
    return (6-4*x+x**2)/(6+2*x)

def example(x):
    return np.exp(-x)
```

```python
x = np.linspace(0,5,1000)
approx1 = z1(x)
approx2 = z2(x)
examp = example(x)
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,examp,label='e^(-x)',color = 'r', linewidth = 3 )
plt.plot(x,approx1,label='Pryblienie Padego 1',color = 'b', linewidth = 3, linestyle='--' )
plt.plot(x,approx2,label='Pryblienie Padego 2',color = 'g', linewidth = 3, linestyle='-.' )
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Porwnanie przyblie Padego i dokadnej funkcji e^(-x)')
plt.savefig('przybizenia_razem.png')
plt.show()
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.plot(x,examp,label='e^(-x)',color = 'r', linewidth = 3 )
plt.plot(x,approx1,label='Pryblienie Padego 1',color = 'b', linewidth = 3, linestyle='--' )
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,examp,label='e^(-x)',color = 'r', linewidth = 3 )
plt.plot(x,approx2,label='Pryblienie Padego 2',color = 'g', linewidth = 3, linestyle='-.' )
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)


plt.suptitle('Porwnanie przyblie Padego i dokadnej funkcji e^(-x)')
plt.savefig('przybizenia_osobno.png')
plt.show()
```

```python
error_1 = np.zeros(len(examp))
error_2 = np.zeros(len(examp))

for i,n in enumerate(approx1):
    error = np.abs(n - examp[i])
    error_1[i] = error

for i,n in enumerate(approx2):
    error = np.abs(n - examp[i])
    error_2[i] = error
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,error_1,label='Bd przyblienia Padego 1',color = 'b', linewidth = 3, linestyle='--' )
plt.plot(x,error_2,label='Bd przyblienia Padego 2',color = 'y', linewidth = 3, linestyle='-.' )
plt.xlabel('x')
plt.ylabel('Bd bezwzgldny')
plt.legend()
plt.grid(True)
plt.savefig('bledy.png')
plt.show()
```

Przyblienia Padego s bardzo dokadne dla $ x \in [0,1] $. Dokadno ta zaczyna zanika dla wikszych $x$. Moemy te zauway, e pierwsze przyblienie Padego ma mniejszy bd dla wikszych $x$ w porwnaniu do drugiego. Mona wic wywnioskowa, e pierwsze przyblienie jest lepszym wyborem.

# Zadanie 2

```python
def dokladnosc():
    epsilon = 1

    while 1 + epsilon != 1:
        epsilon /= 2
    return epsilon 

epsilon = dokladnosc()

print(f'Dokadno maszynowa wynosi: {epsilon}')
```

# Zadanie 5

```python
def w(x):  
    return 6* x**4 + 5* x**3 - 13* x**2 + x + 1
```

```python
xs = np.arange(-10,10,0.0001)

poczatek = time.time()
wyniki = w(xs)
koniec = time.time()
czas = koniec - poczatek

plt.figure(figsize=(15,8))
plt.plot(xs, wyniki, label = 'Warto wielomianu',color = 'r', linewidth = 3 )
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('W(x)')
plt.title('Wartoci wielomianu W(x)')
plt.show()

print(f"Czas wykonania: {czas:.6f} sekundy")
```

# Zadanie 6

```python
def horner(x):
    wspolczynniki = [6, 5, -13, 1, 1] 
    wynik = wspolczynniki[0]
    for i in wspolczynniki[1:]:
        wynik = wynik * x + i
    return wynik
```

```python
xs = np.arange(-10,10,0.0001)

poczatek = time.time()
wyniki_horner = horner(xs) 
koniec = time.time()
czas_horner = koniec - poczatek


plt.figure(figsize=(15, 8))
plt.plot(xs, wyniki_horner, label = 'Warto wielomianu (Horner)', color = 'r', linewidth = 3)
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('W(x)')
plt.title('Wartoci wielomianu W(x) - Schemat Hornera')
plt.show()


print(f"Czas wykonania (Horner): {czas_horner:.6f} sekundy")
```

Czas wykonywania programu, korzystajc ze schematu Hornera, jest duo szybszy ni zwyke obliczanie wartoci wielomianu. Schemat Hornera zajmuje $ 0,006996 $ sekundy, natomiast zwyke obliczanie zajo $ 0.024984 $ sekundy. Oznacza to, e schemat Hornera jest okoo $3,5$ raza szybszy. Wynika to z faktu, e schemat Hornera pozwala ograniczy wykonywanie mnoenia, przez co program zostaje zoptymalizowany.
